```python
df = pd.read_csv("diabetes.csv")
col = df.columns.tolist()
df.head()
```

```python
df.head(10)
drop_ind = [i for i in range(0,10,2)]
print(drop_ind)
df.drop(drop_ind,axis = 0,inplace =True)
df.reset_index(drop = True)
df.head(10)
```

```python
def makeMonkey(args):
    monkey = {}
    
    i = 0
    
    for j in range(0,len(args),3):
        monkey[i] = args[j:j+3]
        i += 1
    
    return monkey

def read_rule(rule):
    rule = makeMonkey(rule)
    print("\nIf ")
    for k,v in rule.items():
        if v[0] == 1:
            mx = max(v[1],v[2])
            mn = min(v[1],v[2])
            print(f"{mn} <= {col[k]} <= {mx} and ")
    print("Then Class = Positive\n")


def delRows(args,sign):
    if(df.shape[0] < 50):
        print("Data set size less then threshold\n")
        return
    
    drop_ind = []
    args = makeMonkey(args)
    score = 0
    index = df.index.tolist()
    for i in index:
        rule_satisfied = True
        for k,v in args.items():
            if v[0] == 1:
                mn = min(v[1],v[2])
                mx = max(v[1],v[2])

                if df.iloc[i][col[k]] < mn or df.iloc[i][col[k]] > mx:
                    rule_satisfied = False
                    break
        if rule_satisfied and df.iloc[i]["Outcome"] == sign:
            score += 1
            drop_ind.append(i)

    df.drop(drop_ind,axis = 0,inplace = True)
    return score
```

```python
for i in range(0,len(args)):
    print(df.index.tolist())
    print(f"Size : {df.shape[0]}\n")
    read_rule(args[i])
    score = delRows(args[i],0)
    print(f"Score : {score}\n")
```
